   'count': the count of values entered for this variable in the dataset (amt of observations which have a value for this variable)
   'mean': the average of the data for this variable
   'std': the standard deviation of the sample, aka the average distance from the mean for this variable
   'min': the minimum value out of all the values for this variable
   '25%': the first quartile, also known as the 25th percentile. a quarter of values are lower than it
   '50%': the second quartile, also known as the median. half of values are lower than it
   '75%': the third quartile, also known as the 75th percentile. a quarter of values are higher than it
   'max': the maximum value out of all values for this variable
